```python
import instaloader

def collect_instagram_posts(query):
    L = instaloader.Instaloader()

    # Recherche de publications liÃ©es au sujet
    posts = L.get_hashtag_posts(query)

    # Affichage des rÃ©sultats (texte et images)
    for post in posts:
        print("Texte:", post.caption)
        print("Images:")
        for i, image in enumerate(post.get_sidecar_nodes()):
            print(f"Image {i+1}: {image.display_url}")
        print("\n")

# Remplacez 'QUERY' par votre propre hashtag ou terme de recherche
query = 'nature'
collect_instagram_posts(query)
```

```python
import instaloader
from pymongo import MongoClient

def collect_instagram_posts_to_mongodb(query, username, password):
    L = instaloader.Instaloader()

    # Connexion Ã  la base de donnÃ©es MongoDB (assurez-vous que MongoDB est en cours d'exÃ©cution)
    client = MongoClient('localhost', 27017)  # Remplacez localhost et le port par vos propres paramÃ¨tres
    db = client['instagram_posts']
    collection = db[query]

    # Authentification
    L.load_session_from_file(username)  # Si vous avez dÃ©jÃ  une session enregistrÃ©e
    if not L.context.is_logged_in:
        L.context.log_in(username, password)

    # Recherche de publications liÃ©es au sujet
    posts = L.get_hashtag_posts(query)

    # Enregistrement des rÃ©sultats (texte et images) dans la base de donnÃ©es MongoDB
    for post in posts:
        post_data = {
            "text": post.caption,
            "images": [image.display_url for image in post.get_sidecar_nodes()]
        }
        collection.insert_one(post_data)

# Remplacez 'QUERY', 'USERNAME', et 'PASSWORD' par vos propres valeurs
query = 'nature'
username = 'votre_nom_utilisateur'
password = 'votre_mot_de_passe'
collect_instagram_posts_to_mongodb(query, username, password)
```

```python
import facebook
from pymongo import MongoClient

# ParamÃ¨tres de connexion MongoDB
MONGODB_URI = "mongodb://localhost:27017/"
DB_NAME = "facebook_data"
COLLECTION_NAME = "posts"

# ParamÃ¨tres de connexion Facebook Graph API
ACCESS_TOKEN = "YOUR_FACEBOOK_ACCESS_TOKEN"  # Remplacez par votre token d'accÃ¨s

# Initialiser le client MongoDB
client = MongoClient(MONGODB_URI)
db = client[DB_NAME]
collection = db[COLLECTION_NAME]

# Initialiser le client Facebook Graph API
graph = facebook.GraphAPI(access_token=ACCESS_TOKEN, version="3.1")


# Fonction pour collecter des posts par rapport Ã  un sujet donnÃ©
def collect_posts(topic, num_posts=10):
    # Recherche de posts liÃ©s au sujet
    posts = graph.request("/search?q={}&type=post&limit={}".format(topic, num_posts))

    # Enregistrement des posts dans MongoDB
    for post in posts['data']:
        post_id = post['id']
        post_data = graph.get_object(post_id)
        
        # Stockage dans la base de donnÃ©es MongoDB
        collection.insert_one(post_data)

if __name__ == "__main__":
    # Sujet de recherche
    search_topic = "le dÃ©cÃ¨s du prÃ©sident Jacques Chirac"

    # Collecte des posts
    collect_posts(search_topic)

    # Affichage des rÃ©sultats stockÃ©s dans MongoDB
    cursor = collection.find()
    for document in cursor:
        print(document)

    # Fermer la connexion MongoDB
    client.close()
```

```python
import facebook

def collect_facebook_posts(query, access_token):
    graph = facebook.GraphAPI(access_token, version="3.0")
    
    # Recherche de publications liÃ©es au sujet
    results = graph.request("/search?q=" + query + "&type=post")
    
    # Affichage des rÃ©sultats
    for post in results.get('data', []):
        print(post.get('message', 'No message'))

# Remplacez ACCESS_TOKEN par votre propre jeton d'accÃ¨s
access_token = "ACCESS_TOKEN"
query = "nature"
collect_facebook_posts(query, access_token)
```
